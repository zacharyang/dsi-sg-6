```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set_style()
```

```python
# Generate random data # 

X=np.random.normal(loc=8,scale=7,size=2200)
y=np.random.normal(loc=10,scale=7,size=100)
feature_cols=['X_'+str(i) for i in range(1,23)]
df=pd.DataFrame(np.concatenate((y.reshape(100,1),X.reshape(100,22)),axis=1),columns=['y']+feature_cols)

# Plotting 22 variables, so set subplot grid to 4 by 6. # 
nrow=4
ncol=6
f, ax= plt.subplots(nrow,ncol,figsize=(30,20))

# Flatten axis so that you can loop through the list # 
axes=ax.flatten()

# Turn off last two axis in the (3,4) and (3,5) position # 

ax[3,4].axis('off')
ax[3,5].axis('off')

# Loop over tuple of columns + axes objects # 
for x,ax in zip(df.columns[1:],axes):
    sns.regplot(x=x,y='y',data=df,ax=ax)
    ax.set_xlabel(x,fontsize=14)
    
# Set big title # 
f.suptitle('Y vs X',fontsize=26,fontweight='bold')

plt.tight_layout(pad=6)

plt.show()
```
