```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,num=100)
e =np.random.normal(scale=5.0,size=100)
y = x + e

## Outlier ## 
y2=y.copy()
y2[98:]=-5000
print('Correlation without outliers:{}, Correlation with outliers:{}'.format(stats.pearsonr(x,y)[0],stats.pearsonr(x,y2)[0]))
```

```python
pear1=round(stats.pearsonr(x,y)[0],2)
pear2=round(stats.pearsonr(x,y2)[0],2)
rho1=round(stats.spearmanr(x,y)[0],2)
rho2=round(stats.spearmanr(x,y2)[0],2)
```

```python
np.round(stats.pearsonr(x,y)[0],decimals=2)
```

```python
f,ax= plt.subplots(1,1,figsize=(10,10))

plt.scatter(x,y2,color='red',marker='.')
ax.set_title('Pearson='+str(stats.pearsonr(x,y2)[0]) + ', Spearman=' + str(stats.spearmanr(x,y2)[0]),fontsize=16)

plt.show()
```
